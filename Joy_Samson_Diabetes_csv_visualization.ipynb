```python
import json

# Define the path to your .ipynb file
# For demonstration, let's use the one that was in your `data` variable previously
ipynb_file_path = '/content/Machine-Learning-Class/visualisation/gbganalyst-script.ipynb'

try:
    with open(ipynb_file_path, 'r', encoding='utf-8') as f:
        notebook_content = json.load(f)

    print(f"Successfully loaded '{ipynb_file_path}' as JSON.\n")

    # Iterate through the cells and print their type and source
    for i, cell in enumerate(notebook_content.get('cells', [])): # .get() to safely access 'cells'
        print(f"Cell {i+1} (Type: {cell.get('cell_type')}):")
        # 'source' is a list of strings, join them for display
        source_code = ''.join(cell.get('source', []))
        print(source_code[:200] + ('...' if len(source_code) > 200 else '')) # Print first 200 chars
        print("\n" + "-" * 50 + "\n")

except FileNotFoundError:
    print(f"Error: The file '{ipynb_file_path}' was not found.")
except json.JSONDecodeError:
    print(f"Error: Could not decode JSON from '{ipynb_file_path}'. It might not be a valid .ipynb file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the diabetes dataset
diabetes = pd.read_csv("/content/Machine-Learning-Class/data_cleaningML/diabetes.csv")

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=diabetes,
    x="BloodPressure",
    y="BMI",
    palette="red"
)
plt.xlabel("Blood Pressure")
plt.ylabel("BMI")
plt.title("Blood Pressure vs BMI")
plt.show()
```
