```python
# Importing the Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the Dataset to Paandas DataFrame
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rupakroy/sonarcsv")

print("Path to dataset files:", path)
```

Data Collection and Data Preprocessing

```python
import pandas as pd
sonar_data = pd.read_csv(path + '/sonar.csv')
print(sonar_data)
```

```python
# Seperating Data and Label
X = sonar_data.drop(columns='R', axis=1)
Y = sonar_data['R']
print(X)
print(Y)
```

```python
Training and Test Data
```

```python
x_test, x_train, y_test, y_train = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
print(X.shape, x_train.shape, x_test.shape)
```

```python
print(x_train)
print(y_train)
```

Model Training  - Logistics Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Train the logistic Regression Model  with training data
model.fit(x_train, y_train)
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score
# Accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# Accuracy on Test Data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print('Accuracy on test data : ', test_data_accuracy)
```

Making a Predictive System

```python
import numpy as np
input_data = X.iloc[0].values.tolist() # Using the first row of the training data as an example
# Changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)
# Reshape the np array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)
if (prediction[0]=='R'):
  print('The object is a Rock')
else:
  print('The object is a Mine')
```
